## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
QUERY = """
select ConvertedComp
from master
"""
df= pd.read_sql_query(QUERY, conn)
df["ConvertedComp"].plot(kind="hist")
plt.title("Histogram of Converted Compensation")
plt.ylabel("Number of Respondents")
plt.xlabel("Annual Salary (USD)")
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
select Age
from master
"""
df_age= pd.read_sql_query(QUERY, conn)
df["Age"].plot(kind="box", x=df_age["Age"])
plt.ylabel("Age of the respondents")
plt.title("Box plot of Age respondents")



```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
select Age, WorkWeekHrs
from master
"""
df_aw= pd.read_sql_query(QUERY, conn)
plt.scatter(data=df_aw, x= "Age", y = "WorkWeekHrs",alpha=0.4)
plt.title("Scatter plot of Age and WorkWeekHrs")
plt.xlabel("Age")
plt.ylabel("Workweekhrs")



```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import seaborn as sns
QUERY = """
select Age, WorkWeekHrs, CodeRevHrs
from master
"""
df_code= pd.read_sql_query(QUERY, conn)
sns.scatterplot(data=df_code, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', sizes=(15,150), alpha=0.5)
plt.title('WorkWeekHrs and CodeRevHrs By Age') 
plt.xlabel('WorkWeekHrs') 
plt.ylabel('CodeRevHrs')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


### Stacked Charts


```python
# your code goes here
QUERY = """ 
select DatabaseDesireNextYear, count(*) as count 
from DatabaseDesireNextYear 
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) desc limit 5 
""" 

df=pd.read_sql_query(QUERY,conn)  
df.set_index('DatabaseDesireNextYear', inplace=True) 

colors=['yellow', 'yellowgreen', 'red', 'blue', 'green', 'pink']

df['count'].plot(kind='pie', figsize=(10,6), autopct='%1.1f%%', labels=None, colors=colors, pctdistance=1.12)  

plt.legend(labels=df.index, loc='upper right')
plt.title('Top 5 Databases Respondents Wish To Learn') 
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
QUERY = """ 
SELECT LanguageWorkedWith, COUNT(*) as Count
FROM LanguageWorkedWith
group by LanguageWorkedWith
order by count(LanguageWorkedWith) desc limit 5
""" 
df=pd.read_sql_query(QUERY,conn)
df.set_index('LanguageWorkedWith', inplace=True) 
df.plot(kind="bar", figsize=(6,3))
plt.title("Top 5 Most Popular Languages to work with")
plt.ylabel("Number of respondents")
plt.legend().remove()
plt.show()
```

```python
QUERY = """ 
SELECT LanguageDesireNextYear, COUNT(*) as Count
FROM LanguageDesireNextYear
group by LanguageDesireNextYear
order by count(LanguageDesireNextYear) desc limit 5
""" 
df=pd.read_sql_query(QUERY,conn)
df.set_index('LanguageDesireNextYear', inplace=True) 
df.plot(kind="bar", figsize=(6,3), color="red")
plt.title("Top 5 Languages desired to learn next year")
plt.ylabel("Number of respondents")
plt.legend().remove()
plt.show()
```

Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """ 
SELECT WorkWeekHrs, CodeRevHrs, Age
from master
where Age between 30 and 35
""" 
df_stacked=pd.read_sql_query(QUERY,conn)
df_stacked= df_stacked.groupby("Age").median()

df_stacked.plot(kind="bar",figsize=(10,6), stacked=True)
plt.title("Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Those Age 30 to 35")
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """ 
SELECT ConvertedComp, Age
from master
where Age between 45 and 60
""" 

df_comp=pd.read_sql_query(QUERY,conn)
df_comp= df_comp.groupby("Age").median()
df_comp.plot(kind="line", figsize=(10,6))
plt.title("Line Chart of Median ConvertedComp for age between 45 and 60")
plt.ylabel("Annual Salary (USD)")
plt.grid(True)

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """ 
SELECT MainBranch, count(*) as Count
from master
group by MainBranch
order by Count desc
""" 
df=pd.read_sql_query(QUERY,conn)
plt.barh(df["MainBranch"], df["Count"])
plt.xlabel("Number of respondents")
plt.title("MainBranch Distribution")

```
